# **For Multiple Notebooks Clean-Up**

````python
import json

# Define the relative paths
notebook_paths = [
    "git/git_commit_florence2.ipynb",
    "notebooks/Florence2_Inference.ipynb",
    "notebooks/florence2_preliminary_results.ipynb"
]

for path in notebook_paths:
    print(f"Checking...: {path}")

    try:
        with open(path, 'r', encoding='utf-8') as f:
            notebook_data = json.load(f)

        # Check for and delete 'widgets' metadata issues
        if 'widgets' in notebook_data.get('metadata', {}):
            del notebook_data['metadata']['widgets']

            # Save the cleaned notebook back to the file
            with open(path, 'w', encoding='utf-8') as f:
                json.dump(notebook_data, f, indent=1)
            print("---> Fixed! Widget metadata removed.\n")
        else:
            print("---> Clean! No widget metadata found.\n")

    except FileNotFoundError:
        print("---> ERROR: File not found. Double-check the path.\n")
````
